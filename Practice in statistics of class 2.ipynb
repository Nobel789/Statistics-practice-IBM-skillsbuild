```python
#To do the probability test we import random module from python
#then we flips the coin with define number with random.choice function
# then calculate the probability of frequncy of head and tail
import random
```

```python
flips = random.choices(["H", "T"], k=20)
flips
```

```python
# from that calculate frequency, total head by total len
fre_H = flips.count("H") / len(flips)
fre_H
```

```python
# its actualy a good prediction , its not usual ,if we cahnge the flips no like 1000
flips = random.choices(["H", "T"], k=1000)
flips
```

```python
fre_H = flips.count("H") / len(flips)
fre_H
```

```python
#we see its change but close to .50
#Lets check  the tail
fre_T = flips.count("T") / len(flips)
fre_T
```

```python
#Lets do the probability with dice
# we know dice have 6 spaces, so define it and folow the filps and count the frequency
dice_space = [1, 2, 3, 4, 5, 6]
```

```python
rolls = random.choices(dice_space, k=12)
rolls
```

```python
fre1_fours = rolls.count(4) / len(rolls)
fre1_fours
```

```python
fre2_fours = rolls.count(4) / len(rolls)
fre2_fours
```

# Set operations
𝐴 and 𝐵 are events

𝐴∩𝐵
  (A intersection B) is another event, which occurs if both  𝐴 and  𝐵 occur: AND#if both events occurs

𝐴∪𝐵
  (A union B) is another event, which occurs if either of  𝐴 or  𝐵 occur: OR#if either one occurs

𝐴′is the opposite of 𝐴

```python
#Example of die
A={1,2,3,4,5,6}
B={3,4,5}
```

```python
𝐴∩𝐵={3,4,5}
𝐴∪𝐵={1,2,3,4,6}
```

```python
#morgan law
𝑃(𝐴∪𝐵)=𝑃(𝐴)+𝑃(𝐵)−𝑃(𝐴∩𝐵)
```
